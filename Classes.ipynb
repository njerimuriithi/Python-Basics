### <u>Instance vs Class Methods </u>
##### 1.Define a class named BankAccount with:
- account_number
- balance
##### 2.Add methods:
- deposit(amount)
- withdraw(amount)
- display_balance()
##### 3.Add a class variable to store the bank name and a class method to display the bank name.
##### 4.Create two account objects and perform deposits and withdrawals.


```python
class BankAccount:
    def __init__(self,account_number,balance,bank_name):
        self.bank_name =bank_name
        self.account_number =account_number
        self.balance =balance

    def deposit(self,amount):
        self.balance += amount       
        return self.balance
    
    def withdraw(self,amount):
        self.balance -= amount
        return self.balance
    
    def display_balance(self):
        return self.balance
    def display_bankName(self):
        return self.bank_name 
    
    def display_details(self):
        print("Account Number:", self.account_number)
        print("Bank Name:", self.bank_name)
        print("Balance:", self.balance)
```

```python
account1 = BankAccount("9781245",0,"Equity")
account1.deposit(90000)
account1.withdraw(8500)
account1.display_balance()
account1.display_bankName()
account1.display_details()
```

### <u>Encapsulation</u>
- protecting data inside
- create a getter method to access private properties
##### 1.Create a class Car with:
- Private attributes: __make, __model, __year, __speed
- Methods to:
- Get and set make, model, and year.
- Increase and decrease speed.
- Display all car details.
##### 2.Create a Car object, update its attributes, and demonstrate encapsulation.

```python
class car:
    def __init__(self):
        self.__Make =""
        self.__model =""
        self.__year=""
        self.__speed=0
    def set_carattributes(self,make,model,year,speed):
        self.__Make =make 
        self.__model=model
        self.__year =year
        self.__speed=speed 
    
    def get_carattributes(self):
        car_attribute =f"This car is of make:{self.__Make},Model:{self.__model},speed:{self.__speed},year:{self.__year}"
        return car_attribute
     
 
        
```

```python
car = car()
car.set_carattributes("Hyundai","Axela","2021",452)
print(car.get_carattributes())
```

### <u>Methods with Objects as Arguments</u>
##### 1.Create a class Circle with:
- Attributes: radius
- Method area() and circumference().
##### 2.Create a class Cylinder that takes a Circle object and height.
- Method volume() to compute the volume using the circle's area.
##### 3.Demonstrate by creating a Circle object and passing it to the Cylinder object.

```python
class Cicle:
    def __init__(self,radius):
        self.radius =radius
    
    def area(self):
        area=3.14*self.radius*self.radius
        return  area
    
    def circumference(self):
        circum = 2*3.14*self.radius
        return circum
    
class Cylinder(Cicle):
        def __init__(self,height,radius):
            super().__init__(radius) # Super calls the method of the from the parent class
            self.height = height
        def volume(self):            
            return self.area()*self.height
```

```python
Cyl1= Cylinder(7,10)
print(f"area:{Cyl1.area()},circumference:{Cyl1.circumference()},Volume:{Cyl1.volume()}")
```

### <u> Object Relationships (Aggregation)</u>
##### 1.Create a class Author with:
- Attributes: name, nationality
- Method to display author info.
##### 2.Create a class Book with:
- Attributes: title, price, and author (Author object).
- Method to display book details including author info.
##### 3.Create one author and multiple books related to that author.

```python
class Author:
    def __init__(self,name,nationality):
        self.name=name
        self.nationality = nationality
    
    def display_authordetails(self):
        author_info=f"Name:{self.name}, Nationality:{self.nationality}"
        return author_info
    
class Book(Author):
    def __init__(self, name, nationality,title,price):
        super().__init__(name, nationality)
        self.price =price
        self.title=title
    def display_bookDetails(self):
        print("Author info:", self.display_authordetails())
        print("Book Title:", self.title,self.price)
        print("Book Prce:", self.price)
      
        
        
```

```python
Autho1=Author("Miriam Makeba","Kenya")
book1=Book(Autho1.name,Autho1.nationality,"Woman Evolve",9.9990)
book1.display_bookDetails()

book2=Book(Autho1.name,Autho1.nationality,"Stories Of Nairobi Women",19.9990)
book2.display_bookDetails()

```
