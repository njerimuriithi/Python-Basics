3. Write a program that takes 3 numbers and prints the largest one.


```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

if num1 >= num2 and num1 >= num3:
    
    print("Largest number is:", num1)
elif num2 >= num1 and num2 >= num3:
    print("Largest number is:", num2)
else:
    print("Largest number is:", num3)
```

#### Build a grading system:

- Input score (0–100)
- Output grade: A (90+), B (80–89), etc.

```python
score = int(input("Enter The Students Score:"))
print(f"Student Score = {score}")
if score >=90:
    print("A")
elif score >=80 and score <=89:
    print("B")
elif score >=55 and score <80:
    print("C")
elif score >=40 and score<55:
    print("D")
else:
    print("E")
```
