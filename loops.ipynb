2. Use a while loop to print numbers until the user enters stop.


```python
print("Enter Numbers, Type stop to end program")
while True:
    userInput =input("Enter a Number:")
    if userInput =='stop':
        print("Program Stopped")
        break
    else:
        number=int(userInput)
        print(f"You entered No {number}")
```

### <u>Challenge</u>
##### Write a guessing game that asks the user to guess a secret number between 1 and 10.

- Give feedback (too high / too low)
- Use a while loop

```python
print("Guess the Number between 1 and 10")
secret_No =8
while True:
    user_input =int(input("Enter Any Number:"))
    if user_input < secret_No:
        print("too low")

    elif user_input >secret_No:
        print("Too high")
      
    elif user_input <1 or user_input >10:
        print("Number out of expected range")
    else:
        print("Correct Guess")
        break
        
        

```
