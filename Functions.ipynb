### <u>Challenge</u>
#### Write a calculator function:

- Takes two numbers and an operation (+, -, *, /)
- Returns the result

```python
print("Enter Two Numbers")

def calculator():
    num1 =int(input("Enter num1:"))
    num2=int(input("enter num2:"))
    operation =input("Enter operation (+, -, *, /): ")
    if operation =="*":
        print(num1*num2)
    elif operation =="+":
        print(num1+num2)
    elif operation =="-":
        print(num1-num2)
    else:
        print("C")

calculator()
            
```
